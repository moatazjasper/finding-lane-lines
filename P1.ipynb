## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def convert_hls(image):
    # Convert to HLS Color Space
    return cv2.cvtColor(image, cv2.COLOR_RGB2HLS)

def select_white_yellow(image):
    # Convert to HLS Color Space
    converted = convert_hls(image)
    
    # Define white color mask
    lower = np.uint8([  0, 200,   0])
    upper = np.uint8([255, 255, 255])
    white_mask = cv2.inRange(converted, lower, upper)
    
    # Define yellow color mask
    lower = np.uint8([ 10,   0, 100])
    upper = np.uint8([ 40, 255, 255])
    yellow_mask = cv2.inRange(converted, lower, upper)
    
    # Combine both masks
    mask = cv2.bitwise_or(white_mask, yellow_mask)
    
    return cv2.bitwise_and(image, image, mask = mask)

def grayscale(image):
    # Select only white and yellow pixels as a form of color selection
    wyImage = select_white_yellow(image)
    
    # Convert resulting image to grayscale
    return cv2.cvtColor(wyImage, cv2.COLOR_RGB2GRAY)
    
def canny(img, low_threshold, high_threshold):
    # Apply Canny Transform for Edge Detection
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def get_lines_slope_intercept(lines):
    """
    Line Equation: y=mx+b .
    Calculate slope by: (y2-y1)/(x2-x1)
    Calculate intercept from the line equation.
    """
    lines_slope_intercept = np.zeros(shape=(len(lines),2))
    for index,line in enumerate(lines):        
        for x1,y1,x2,y2 in line:
            slope = (y2-y1)/(x2-x1)
            intercept = y1 - x1 * slope
            lines_slope_intercept[index]=[slope,intercept]
    return lines_slope_intercept

def draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """
    Modified Interface...
    """
    # Getting the dimensions of the image
    x_size = img.shape[1]
    y_size = img.shape[0]
    
    # Calling method for calculating slopes and intercepts
    lines_slope_intercept = get_lines_slope_intercept(lines)
    
    # Max slope will be for the left lane line
    max_slope_line = lines_slope_intercept[lines_slope_intercept.argmax(axis=0)[0]]
    # Min slope will be for the right lane line
    min_slope_line = lines_slope_intercept[lines_slope_intercept.argmin(axis=0)[0]]

    # Arrays for clustering slopes and intercepts for left and right lane lines
    left_slopes = []
    left_intercepts = []
    right_slopes = []
    right_intercepts = []
    
    # Gets slopes and intercepts of lines similar to:
    # the lines with the maximum slope (left lane)
    # and the lines with minimum slope (right lane),
    # within a tolerance percentage (tol_per)
    tol_per = 0.15
    for line in lines_slope_intercept:
        if abs(line[0] - max_slope_line[0]) < tol_per and abs(line[1] - max_slope_line[1]) < (tol_per * x_size):
            left_slopes.append(line[0])
            left_intercepts.append(line[1])
        elif abs(line[0] - min_slope_line[0]) < tol_per and abs(line[1] - min_slope_line[1]) < (tol_per * x_size):
            right_slopes.append(line[0])
            right_intercepts.append(line[1])

    # Left and right lines are averages of slopes and intercepts.
    # Extrapolate lines to edges and center, roughly.
    new_lines = np.zeros(shape=(1,2,4), dtype=np.int32)
    if len(left_slopes) > 0:
        # Create left average line as an array of [average of slopes, average of intercepts]
        left_line = [sum(left_slopes)/len(left_slopes),sum(left_intercepts)/len(left_intercepts)]
        
        # Define the x dimension of the two points of the average left line
        left_bottom_x = (y_size - left_line[1])/left_line[0]
        left_top_x = (y_size*0.575 - left_line[1])/left_line[0]
        
        # Create left lane line
        if (left_bottom_x >= 0):
            new_lines[0][0] =[left_bottom_x,y_size,left_top_x,y_size*.575]

    if len(right_slopes) > 0:
        # Create right average line as an array of [average of slopes, average of intercepts]
        right_line = [sum(right_slopes)/len(right_slopes),sum(right_intercepts)/len(right_intercepts)]
        
        # Define the x dimension of the two points of the average right line
        right_bottom_x = (y_size - right_line[1])/right_line[0]
        right_top_x = (y_size*0.575 - right_line[1])/right_line[0]
        
        # Create right lane line
        if (right_bottom_x <= x_size):
            new_lines[0][1]=[right_bottom_x,y_size,right_top_x,y_size*.575]

    for line in new_lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
def pipeline_interface(image):
    #1. Convert Image to Greyscale
    grey = grayscale(image)

    #2. Define Kernel size and Blur for Gaussian blurring
    kernel_size = 5
    blur_grey = gaussian_blur(grey,kernel_size)

    #3. Define Thresholds for Canny Edge Detection
    low_threshold = 75
    high_threshold = 150
    canny_edges = canny(blur_grey, low_threshold, high_threshold)

    #4. Define and draw mask polygon
    yShape = image.shape[0]
    xShape = image.shape[1]
    bottom_left  = [xShape*0., yShape]
    top_left     = [xShape*0.475 , yShape*0.575]
    bottom_right = [xShape, yShape]
    top_right    = [xShape*0.525 , yShape*0.575]
    vertices = np.array([[bottom_left, top_left, top_right, bottom_right]], dtype=np.int32)
    masked_edges = region_of_interest(canny_edges, vertices)

    #5. Define Hough Transform Parameters and Apply transformation on Canny o/p
    rho = 3                       # distance resolution in pixels of the Hough grid
    theta = np.pi/180             # angular resolution in radians of the Hough grid
    threshold = 70                # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 70          # minimum number of pixels making up a line
    max_line_gap = 250             # maximum gap in pixels between connectable line segments
    line_image = np.copy(image)*0 # creating a blank to draw lines on

    hough_transform_lines = hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)

    6. # Draw the lines on the edge image
    new_img = weighted_img(hough_transform_lines, image)
    return new_img
    
for index, img in enumerate(os.listdir("test_images/")):
    image = mpimg.imread('test_images/' + img)
    new_img = pipeline_interface(image)
    plt.figure()
    plt.imshow(new_img)
````
